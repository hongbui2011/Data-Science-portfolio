# Project Proposal
Since credit card default can be perceived as financial fraud, it is critical that we precisely identify or early detect possible default cases in order to prevent the losses. The aim of this project is to predict the chances of customers being unable to repay their credit card debt next month using different Machine Learning models 

# Data Summary

The dataset used in this project was obtained from UCI Machine Learning Repository https://archive.ics.uci.edu/ml/datasets/default+of+credit+card+clients

The dataset contains 30,000 observations with 25 variables.

### Data Dictionary





**ID**: Unique ID of each client

**LIMIT_BAL**: Amount of the given credit (NT dollar): it includes both the individual consumer credit and his/her family (supplementary) credit.

**SEX**: Gender (1 = male; 2 = female).

**EDUCATION**: Education (1 = graduate school; 2 = university; 3 = high school; 4 = others).

**MARRIAGE**: Marital status (1 = married; 2 = single; 3 = others).

**AGE**: Age (year).

**PAY_0**: Repayment status in September, 2005 (-1=pay duly, 1=payment delay for one month, 2=payment delay for two months, ... 8=payment delay for eight months, 9=payment delay for nine months and above)

**PAY_2**: Repayment status in August, 2005 (scale same as above)

**PAY_3**: Repayment status in July, 2005 (scale same as above)

**PAY_4**: Repayment status in June, 2005 (scale same as above)

**PAY_5**: Repayment status in May, 2005 (scale same as above)

**PAY_6**: Repayment status in April, 2005 (scale same as above)

**BILL_AMT1**: Amount of bill statement in September, 2005 (NT dollar)

**BILL_AMT2**: Amount of bill statement in August, 2005 (NT dollar)

**BILL_AMT3**: Amount of bill statement in July, 2005 (NT dollar)

**BILL_AMT4**: Amount of bill statement in June, 2005 (NT dollar)

**BILL_AMT5**: Amount of bill statement in May, 2005 (NT dollar)

**BILL_AMT6**: Amount of bill statement in April, 2005 (NT dollar)

**PAY_AMT1**: Amount of previous payment in September, 2005 (NT dollar)

**PAY_AMT2**: Amount of previous payment in August, 2005 (NT dollar)

**PAY_AMT3**: Amount of previous payment in July, 2005 (NT dollar)

**PAY_AMT4**: Amount of previous payment in June, 2005 (NT dollar)

**PAY_AMT5**: Amount of previous payment in May, 2005 (NT dollar)

**PAY_AMT6**: Amount of previous payment in April, 2005 (NT dollar)

**default payment next month**: default payment (Yes = 1, No = 0)

# Data Preprocessing

### 1. Check missing data

```python
total = creditcard.isna().sum().sort_values(ascending = False)
percent = (total/creditcard.isnull().count()*100).sort_values(ascending = False)
pd.concat([total, percent], axis=1, keys=['Total', 'Percent']).transpose()
```

No missing values detected

### 2. Check duplicated data

```python
creditcard.duplicated().sum()
```

No duplicated data detected

### 3. Feature Engineering

```python
# Creating dummy varaibles for "SEX"
print(creditcard['SEX'].value_counts())

sex_dummies = pd.get_dummies(creditcard['SEX'])
creditcard['SEX_MALE'] = sex_dummies[1]
creditcard['SEX_FEMALE'] = sex_dummies[2]
creditcard.drop(columns=['SEX'], inplace=True)
```

```python
# Creating dummy variables for "MARRIAGE", and merge undefined observations into 4 which is 'Others'
print(creditcard['MARRIAGE'].value_counts())

marriage_dummies = pd.get_dummies(creditcard['MARRIAGE'])
creditcard['MARRIAGE_MARRIED'] = marriage_dummies[2]
creditcard['MARRIAGE_SINGLE'] = marriage_dummies[1]
creditcard['MARRIAGE_OTHER'] = marriage_dummies[0] + marriage_dummies[3]
creditcard.drop(columns=['MARRIAGE'], inplace=True)
```

```python
# ID column has no predictive value. So we cam drop this column for better training model purposes.
del creditcard['ID']
```

# Predictive Models

### 1. Spliting training and testing dataset

```python
creditcard.columns
```

```python
feature_cols = np.array(['LIMIT_BAL', 'EDUCATION', 'AGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4',
       'PAY_5', 'PAY_6', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4',
       'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3',
       'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6', 'SEX_MALE',
       'SEX_FEMALE', 'MARRIAGE_MARRIED', 'MARRIAGE_SINGLE', 'MARRIAGE_OTHER'])
```

```python
X = creditcard[feature_cols]
y = creditcard['payment_default']

#  splitting the data into 80% train and 20% test datasets 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=142, stratify = y)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### 2.Training data on different models

```python
models = {
    LogisticRegression(): "Logistic Regression",
    KNeighborsClassifier(): "K-Nearest Neighbors",
    RandomForestClassifier(): "Random Forest"
}

for model in models.keys():
    model.fit(X_train, y_train)
```

###  3. Testing models

```python
# Tesing model and calculaing the score of each model
for model, model_name in models.items():
    print(model_name + " accuracy score"  + ": {:.3f}".format(model.score(X_test, y_test)))
```

It appears that Random Forest gives the best performance with an accuracy score of 0.814, higher than other models. KNN is the most underperformed model with the lowest accuracy score

```python
# Calculate the Mean Squared Error of each model
for model, model_name in models.items():
    predict = model.predict(X_test)
    MSE = mean_squared_error(y_test, predict)
    print(f'Mean Squared Error of {model_name} = {MSE:.3f}')
    plt.barh(model_name, MSE , color ='indianred')
    plt.title(" MSE of each model", fontsize = 14)
```

It is also clear that Random Forest gives the lowest MSE compared to other models and KNN has the highest error. However, I will try to make improvements on both Random Forest and K-Nearest Model and see how they perform afterward

### 4. Tuning Random Forest 

#### Tuning hyperparameters

Since Grid Search will take a long time to obtain the best parameters for our model, it would be better to use Randomized Search in this case to have an idea of what our optimal hyperparameters should be. 

```python
from sklearn.model_selection import RandomizedSearchCV
rfc=RandomForestClassifier(random_state=142)
```

```python
random_grid = { 
    'n_estimators': [150, 300],
    'max_features': ['auto', 'sqrt'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
```

```python
ransearch_rfc = RandomizedSearchCV(estimator=rfc, param_distributions=random_grid, cv= 5)
ransearch_rfc.fit(X_train, y_train)
```

```python
ransearch_rfc.best_params_
```

#### Feature Selection

```python
estimator = RandomForestClassifier(random_state = 142,
                                   n_estimators = 150, 
                                   max_features = 'sqrt', 
                                   max_depth = 7, 
                                   criterion = 'entropy')
selector = RFE(estimator, 15) #choosing the best 20 features
selector = selector.fit(X_train, y_train)

supp = selector.get_support()
print("Selected features:", feature_cols[supp])

# test the model
predicted = selector.predict(X_test)
print("MSE:", mean_squared_error(y_test, predicted))
print("Accuracy:", accuracy_score(y_test, predicted))
```

 The accuracy and MSE of the model are slightly improved as a result of hyperparameter tuning and feature selection

```python
# Confusion matrix after tuning model
cf_matrix = confusion_matrix(y_test, predicted)
cf_matrix
```

```python
plt.figure(figsize=(10,10))
sns.set(font_scale = 1.5)

ax = sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, 
            fmt='.2%', cmap='BuPu')

ax.set_title('Random Forests Confusion Matrix\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['No Default','Default'])
ax.yaxis.set_ticklabels(['No Default','Default'])

## Display the visualization of the Confusion Matrix.
plt.show()
```

### 5. Tuning K-Nearest Neighbors

#### Tuning hyperparameters

```python
knn = KNeighborsClassifier()

leaf_size = list(range(1,50))
n_neighbors = list(range(1,40))
weights = ['uniform', 'distance']
p = [1,2]

knn_random_grid = dict(leaf_size = leaf_size,
                      n_neighbors = n_neighbors,
                      weights = weights,
                      p = p)
```

```python
ransearch_knn = RandomizedSearchCV(estimator=knn, param_distributions=knn_random_grid, cv= 5)
ransearch_knn.fit(X_train, y_train)
```

```python
ransearch_knn.best_params_
```

#### Feature Selection

```python
feature_score = pd.DataFrame(columns=["Feature", "Accuracy Score"])
for feature in X_train.columns:
    knn = KNeighborsClassifier(n_neighbors = 28)
    knn.fit(X_train[[feature]], y_train)
    knn.score(X_test[[feature]], y_test)
    feature_score = feature_score.append({"Feature": feature, 
                                          "Accuracy Score": knn.score(X_test[[feature]], y_test)},
                                           ignore_index=True)
```

```python
# Ploting feature importance 
feature_score = feature_score.sort_values("Accuracy Score", ascending = False)
plt.figure(figsize=(10,10))
ax = sns.barplot(data=feature_score,
                 x="Accuracy Score",
                 y="Feature", 
                 palette = "gray")
ax.set_title("Accuracy Score for Each Feature")
plt.show()
```

```python
best_features = list(feature_score[feature_score["Accuracy Score"] > 0.8].copy()["Feature"])

best_knn = KNeighborsClassifier(n_neighbors = 28,
                          weights = 'uniform',
                          p = 2,
                          leaf_size = 46)

best_knn.fit(X_train[best_features], y_train)

best_knn_predict = best_knn.predict(X_test[best_features])

knn_testing_accuracy = best_knn.score(X_test[best_features], y_test)
print("Accuracy on testing set:", knn_testing_accuracy)
print("MSE:", mean_squared_error(y_test, best_knn_predict))
```

```python
# Confusion matrix after tuning model
knn_matrix = confusion_matrix(y_test, best_knn_predict)
knn_matrix
```

```python
plt.figure(figsize=(10,10))
sns.set(font_scale = 1.5)

ax = sns.heatmap(knn_matrix/np.sum(knn_matrix), annot=True, 
            fmt='.2%', cmap='BuPu')

ax.set_title('K-Nearest Neighbors Confusion Matrix\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['No Default','Default'])
ax.yaxis.set_ticklabels(['No Default','Default'])

## Display the visualization of the Confusion Matrix.
plt.show()
```

The accuracy and MSE of KNN model significantly improve after the process of tuning hyperparameters and feature selection

# Conclusion

```python
models = {
    'Model': ['Random Forests', 'K-Nearest Neighbor'],
    'Accuracy': [accuracy_score(y_test, predicted), knn_testing_accuracy],
    
    'MSE': [mean_squared_error(y_test, predicted), mean_squared_error(y_test, best_knn_predict)],
    
}

result = pd.DataFrame(models)
result
```
